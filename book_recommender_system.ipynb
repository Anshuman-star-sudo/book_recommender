## popularity based recommender

```python
ratings_with_name = ratings.merge(books,on=['ISBN'])
```

```python
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df
```

```python
num_rating_df.rename(columns={'Book-Rating':'num_rating'} , inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
```

```python
popular_df
```

```python
popular_df = popular_df[popular_df['num_rating']>=250].sort_values('avg_rating' , ascending=False)
```

## collaborative filtering

```python
x  = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200 
```

```python
bookish_users = x[x].index 
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(bookish_users)]
```

```python
filtered_rating # users who have rated more than 200 books
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
```

```python
famous_books = y[y].index
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings # books which have more than 50 ratings and are rated by users who have rated more than 200 books
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt # users as column and books titles as rows with grids as their respective rating by that user
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt) # calculates euclidean distance of each book from all other books 
```

```python
similarity_score.shape
```

```python
def recommended(Book_name):
    # index fetch
    index = np.where(pt.index==Book_name)[0][0]
    # getting 5 similar items 
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1] , reverse=True)[1:6]

    data = []

    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)
    return data
```
